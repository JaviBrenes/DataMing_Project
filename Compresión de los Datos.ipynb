Recoleccion de los Datos (carga de datos)

```python
import pandas as pd
url = 'https://github.com/JaviBrenes/dataset/blob/main/Mall_Customer.xlsx?raw=true'
df = pd.read_excel(url)
total_datos = len(df)
print("Cantidad de datos: ", total_datos )

```

Descripcion de los datos

```python
print(df.head())

min_age = df['Age'].min()
max_age = df['Age'].max()
print("\n")

# rango de edades
print("El rango de edad es de", min_age, "a", max_age, "años.")

# Annual Income range
min_income = df['Annual Income (k$)'].min()
max_income = df['Annual Income (k$)'].max()
print("El rango de Annual Income es de", min_income, "a", max_income, "k$.")

```

Exploracion de Datos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear grafico de los ingresos anuales de cada cliente
sns.displot(df['Annual Income (k$)'], kde=True)
plt.title('Distribución de Annual Income')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Frequency')
plt.show()

# contador de cada genero
print("\n")
gender_counts = df['Gender'].value_counts()

# crear grafico circular
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Porcentaje de Género')
plt.show()

# Crear un hsitograma para cada Spending Score
print("\n")
sns.displot(df['Spending Score (1-100)'], kde=True)
plt.title('Distribución de Spending Score')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Frequency')
plt.show()
```
