Cargar la Informacion

```python
import pandas as pd
url = 'https://github.com/JaviBrenes/dataset/blob/main/Mall_Customer.xlsx?raw=true'
df = pd.read_excel(url)
```

Limpieza de datos

```python
# Limpieza de Datos: Mostrar información básica del dataset
print("\nInformación del dataset:")
print(df.info())

# Limpieza de Datos: Comprobar valores nulos
print("\nValores nulos en el dataset:")
print(df.isnull().sum())

# Limpieza de Datos: Eliminación de filas duplicadas (si existen)
df.drop_duplicates(inplace=True)

```

Formateo de Datos y almacenamiento en una base de datos Sqlite

```python
import sqlite3

# Conexión a la base de datos SQLite
conexion = sqlite3.connect('Mall_Customer.db')

cursor = conexion.cursor()

cursor.execute("""
CREATE TABLE clientes (
    CustomerID INTEGER PRIMARY KEY,
    Gender TEXT,
    Age INTEGER,
    Annual_Income INTEGER,
    Spending_Score INTEGER
);
""")

# Convierte el DataFrame a una lista de registros
datos = df.values.tolist()

# Inserta cada registro en la tabla "clientes"
cursor.executemany("""
INSERT INTO clientes (CustomerID, Gender, Age, Annual_Income, Spending_Score)
VALUES (?, ?, ?, ?, ?)
""", datos)

conexion.commit()
conexion.close()
```
