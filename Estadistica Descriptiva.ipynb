Cargar Datos


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt


# Conectar a la base de datos SQLite
conn = sqlite3.connect('Mall_Customer.db')

data = pd.read_sql("SELECT * FROM clientes", conn)
```

Analisis 1. Grupo Jovenes ( 18 a 30 años)

*   Separados por genero, ingresos anuales y puntuacion de gasto



```python
# Filtrar datos por grupos de edad
jovenes = data[(data['Age'] >= 18) & (data['Age'] <= 30)]
adultos = data[(data['Age'] > 30) & (data['Age'] <= 50)]
adultos_mayores = data[data['Age'] > 50]

# Análisis estadístico descriptivo para jóvenes, separados por género
jovenes_descriptive = jovenes.groupby('Gender').agg({
    'CustomerID': 'count',
    'Annual_Income': ['mean', 'std', 'min', 'max'],
    'Spending_Score': ['mean', 'std', 'min', 'max']
}).round(2)

print("\nEstadísticas descriptivas para jóvenes (18-30 años):\n")
print(jovenes_descriptive)
print("\n")
```

Analisis 2. Grupo Adulto( 31 a 50 años )

*   Separados por genero, ingresos anuales y puntuacion de gastos



```python
# Análisis estadístico descriptivo para adultos, separados por género
adultos_descriptive = adultos.groupby('Gender').agg({
    'CustomerID': 'count',
    'Annual_Income': ['mean', 'std', 'min', 'max'],
    'Spending_Score': ['mean', 'std', 'min', 'max']
}).round(2)

print("\nEstadísticas descriptivas para adultos (31-50 años):\n")
print(adultos_descriptive)
```

Analisi 3 . Grupo Adulto Mayor (mayor a 50 años)

*   Separados por genero, ingresos anuales y puntuacion de gastos



```python
# Análisis estadístico descriptivo para adultos mayores, separados por género
adultos_mayores_descriptive = adultos_mayores.groupby('Gender').agg({
    'CustomerID': 'count',
    'Annual_Income': ['mean', 'std', 'min', 'max'],
    'Spending_Score': ['mean', 'std', 'min', 'max']
}).round(2)

print("\nEstadísticas descriptivas para adultos mayores (mayores de 50 años):\n")
print(adultos_mayores_descriptive)
```
